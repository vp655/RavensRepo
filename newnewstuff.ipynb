```python
# --- NFL Play Animation with WR + CB Highlights (robust) ---
# Requirements: pip install pandas matplotlib numpy pillow
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation

# ================== CONFIG ==================
GAMES_CSV = "data/games.csv"
PLAYS_CSV = "data/plays.csv"
WEEK_GLOB = "data/week*.csv"

GAME_ID = 2018090600
PLAY_ID = 75
str_GAME_ID = f"{GAME_ID}_{PLAY_ID}"
OUT_GIF = f"animations/{str_GAME_ID}_animation.gif"

# ---- Highlight config ----
# Pick one key present in your tracking: "nflId", "jerseyNumber", or "displayName"
HIGHLIGHT_WR_BY = "nflId"
HIGHLIGHT_WR_VAL = 2495454     # <-- WR id/value to highlight (black)

HIGHLIGHT_CB_BY = "nflId"
HIGHLIGHT_CB_VAL = 2555383     # <-- CB id/value to highlight (red)

# Team colors (primary, secondary)
TEAM_COLORS = {
    "ARI": ("#97233F", "#000000"), "ATL": ("#A71930", "#000000"),
    "BAL": ("#241773", "#9E7C0C"), "BUF": ("#00338D", "#C60C30"),
    "CAR": ("#0085CA", "#101820"), "CHI": ("#0B162A", "#C83803"),
    "CIN": ("#FB4F14", "#000000"), "CLE": ("#311D00", "#FF3C00"),
    "DAL": ("#041E42", "#869397"), "DEN": ("#FB4F14", "#002244"),
    "DET": ("#0076B6", "#B0B7BC"), "GB":  ("#203731", "#FFB612"),
    "HOU": ("#03202F", "#A71930"), "IND": ("#002C5F", "#A2AAAD"),
    "JAX": ("#006778", "#9F792C"), "KC":  ("#E31837", "#FFB81C"),
    "LA":  ("#0080C6", "#FFC20E"), "LAR": ("#003594", "#FFA300"),
    "LV":  ("#000000", "#A5ACAF"), "MIA": ("#008E97", "#FC4C02"),
    "MIN": ("#4F2683", "#FFC62F"), "NE":  ("#002244", "#C60C30"),
    "NO":  ("#D3BC8D", "#101820"), "NYG": ("#0B2265", "#A71930"),
    "NYJ": ("#125740", "#FFFFFF"), "PHI": ("#004C54", "#A5ACAF"),
    "PIT": ("#FFB612", "#101820"), "SEA": ("#002244", "#69BE28"),
    "SF":  ("#AA0000", "#B3995D"), "TB":  ("#D50A0A", "#FF7900"),
    "TEN": ("#0C2340", "#4B92DB"), "WAS": ("#5A1414", "#FFB612"),
}

# ================== LOAD DATA ==================
if not os.path.exists(GAMES_CSV) or not os.path.exists(PLAYS_CSV):
    raise FileNotFoundError("games.csv and plays.csv must be present in 'data/'.")

games = pd.read_csv(GAMES_CSV)
plays = pd.read_csv(PLAYS_CSV)

game_row = games.loc[games["gameId"] == GAME_ID]
if game_row.empty:
    raise ValueError(f"gameId {GAME_ID} not found in games.csv")
game_row = game_row.iloc[0]
actual_week = int(game_row["week"])

week_file_needed = f"data/week{actual_week}.csv"
if not os.path.exists(week_file_needed):
    week_files = sorted(glob.glob(WEEK_GLOB))
    tracking, selected_week = None, None
    for wf in week_files:
        try:
            tmp = pd.read_csv(wf, usecols=["gameId","playId"])
            mask = (pd.to_numeric(tmp["gameId"], errors="coerce") == GAME_ID) & \
                   (pd.to_numeric(tmp["playId"], errors="coerce") == PLAY_ID)
            if mask.any():
                tracking = pd.read_csv(wf)
                selected_week = wf
                break
        except Exception:
            continue
    if tracking is None:
        raise ValueError(f"Play {GAME_ID}/{PLAY_ID} not found in any data/week*.csv")
else:
    tracking = pd.read_csv(week_file_needed)
    selected_week = week_file_needed

# Subset to play
track = tracking[(pd.to_numeric(tracking["gameId"], errors="coerce") == GAME_ID) &
                 (pd.to_numeric(tracking["playId"], errors="coerce") == PLAY_ID)].copy()

# ================== PREP & CHECKS ==================
required_cols = {"gameId","playId","frameId","x","y","team","s","dir"}
missing = required_cols - set(track.columns)
if missing:
    raise ValueError(f"Tracking {selected_week} missing required columns: {missing}")

for c in ["gameId","playId","frameId"]:
    track[c] = pd.to_numeric(track[c], errors="coerce")

frame_ids = track["frameId"].dropna().astype(int).sort_values().unique().tolist()
if len(frame_ids) == 0:
    raise ValueError("No frames found for this play.")
print(f"Frames found: {len(frame_ids)}  (first={frame_ids[0]}, last={frame_ids[-1]})")

home_abbr = str(game_row["homeTeamAbbr"])
away_abbr = str(game_row["visitorTeamAbbr"])
home_1, home_2 = TEAM_COLORS.get(home_abbr, ("#2b8cbe", "#045a8d"))
away_1, away_2 = TEAM_COLORS.get(away_abbr, ("#de2d26", "#a50f15"))

play_dir = str(track.iloc[0].get("playDirection", "right")).lower()
abs_yl = float(plays.loc[(plays["gameId"] == GAME_ID) & (plays["playId"] == PLAY_ID),
                         "absoluteYardlineNumber"].iloc[0])
yards_to_go = float(plays.loc[(plays["gameId"] == GAME_ID) & (plays["playId"] == PLAY_ID),
                              "yardsToGo"].iloc[0])

if play_dir == "left":
    line_of_scrimmage = abs_yl
    to_go_line = line_of_scrimmage - yards_to_go
else:
    line_of_scrimmage = 100 - abs_yl  # rightward
    to_go_line = line_of_scrimmage + yards_to_go

desc_col = next((c for c in plays.columns if c.lower() == "playdescription"), None)
title_txt = str(plays.loc[(plays["gameId"] == GAME_ID) & (plays["playId"] == PLAY_ID), desc_col].iloc[0]) \
            if desc_col else f"Play {GAME_ID}-{PLAY_ID}"

# Keep needed columns; add velocity
need_cols = ["x","y","s","dir","event","displayName","jerseyNumber","frameId","team","nflId"]
track = track[[c for c in need_cols if c in track.columns]].copy()

rad = np.deg2rad(pd.to_numeric(track["dir"], errors="coerce"))
speed = pd.to_numeric(track["s"], errors="coerce").fillna(0.0)
track["v_x"] = np.sin(rad) * speed
track["v_y"] = np.cos(rad) * speed

# Normalize highlight columns
for col in {HIGHLIGHT_WR_BY, HIGHLIGHT_CB_BY} - {None}:
    if col and col in track.columns:
        if col in ["nflId", "jerseyNumber", "frameId"]:
            track[col] = pd.to_numeric(track[col], errors="coerce")
        else:
            track[col] = track[col].astype(str)

# Map team labels 'home'/'away' to abbreviations
def map_team(t):
    t = str(t)
    if t == "home": return home_abbr
    if t == "away": return away_abbr
    return t
track["team"] = track["team"].astype(str).apply(map_team)

home_df = track[track["team"] == home_abbr]
away_df = track[track["team"] == away_abbr]
ball_df = track[track["team"] == "football"]

# ================== FIELD DRAWING ==================
FIELD_W = 120.0
FIELD_H = 160.0/3.0
ARROW_SCALE = 1.0

def draw_field(ax, field_color="#ffffff", line_color="#212529", number_color="#adb5bd"):
    ax.set_xlim(0, FIELD_W); ax.set_ylim(0, FIELD_H); ax.set_aspect("equal")
    ax.set_facecolor(field_color); ax.set_xticks([]); ax.set_yticks([])
    for s in ax.spines.values(): s.set_visible(False)

    ax.plot([0, FIELD_W], [0, 0], color=line_color, lw=1)
    ax.plot([0, FIELD_W], [FIELD_H, FIELD_H], color=line_color, lw=1)
    ax.plot([0, 0], [0, FIELD_H], color=line_color, lw=1)
    ax.plot([FIELD_W, FIELD_W], [0, FIELD_H], color=line_color, lw=1)
    for x in np.arange(10, 111, 5):
        ax.plot([x, x], [0, FIELD_H], color=line_color, lw=0.5)

    xs = np.arange(10, 111, 1)
    hash_y1 = (160/6 + 18.5/6)
    hash_y2 = (160/6 - 18.5/6)
    for x in xs:
        ax.plot([x, x], [0, 1], color=line_color, lw=0.5)
        ax.plot([x, x], [FIELD_H-1, FIELD_H], color=line_color, lw=0.5)
        ax.plot([x, x], [hash_y1, hash_y1+1], color=line_color, lw=0.5)
        ax.plot([x, x], [hash_y2-1, hash_y2], color=line_color, lw=0.5)

    nums = list(range(10, 60, 10)) + list(range(40, 0, -10))
    for i, x in enumerate(range(20, 101, 10)):
        ax.text(x, 12, str(nums[i]), color=number_color, ha="center", va="center", fontsize=12)
        ax.text(x, FIELD_H-12, str(nums[i]), color=number_color, ha="center", va="center",
                fontsize=12, rotation=180)

# ================== ANIMATION ==================
fig, ax = plt.subplots(figsize=(12, 6))
draw_field(ax)
ax.plot([line_of_scrimmage, line_of_scrimmage], [0, FIELD_H], color="#0d41e1", lw=2)
ax.plot([to_go_line, to_go_line], [0, FIELD_H], color="#f9c80e", lw=2)

# Persistent artists (teams + ball + title)
away_pts = ax.scatter([], [], s=250, facecolors="#f8f9fa", edgecolors=away_2, linewidths=2, zorder=3)
home_pts = ax.scatter([], [], s=250, facecolors=home_1, edgecolors=home_2, linewidths=2, zorder=3)
(ball_pt,) = ax.plot([], [], "o", ms=8, mfc="#935e38", mec="#d9d9d9", mew=1.5, zorder=4)
title = ax.set_title(title_txt, fontsize=12, weight="bold")

# Highlighted WR (black) + label + trail
wr_pt = ax.scatter([], [], s=360, facecolors="#000000", edgecolors="#ffffff", linewidths=2.5, zorder=6)
wr_label = ax.text(0, 0, "", color="#ffffff", fontsize=9, ha="center", va="bottom", zorder=7)
wr_trail_x, wr_trail_y = [], []
wr_trail, = ax.plot([], [], lw=2, color="#000000", alpha=0.5, zorder=5.5)

# Highlighted CB (red) + label + trail
cb_pt = ax.scatter([], [], s=360, facecolors="#cc0000", edgecolors="#ffffff", linewidths=2.5, zorder=6)
cb_label = ax.text(0, 0, "", color="#ffffff", fontsize=9, ha="center", va="bottom", zorder=7)
cb_trail_x, cb_trail_y = [], []
cb_trail, = ax.plot([], [], lw=2, color="#cc0000", alpha=0.5, zorder=5.5)

# Dynamic artists (arrows & jersey texts)
jersey_texts = []
arrow_artists = []

def init():
    away_pts.set_offsets(np.empty((0, 2)))
    home_pts.set_offsets(np.empty((0, 2)))
    ball_pt.set_data([], [])
    wr_pt.set_offsets(np.empty((0, 2))); wr_label.set_text(""); wr_trail.set_data([], [])
    cb_pt.set_offsets(np.empty((0, 2))); cb_label.set_text(""); cb_trail.set_data([], [])
    return [away_pts, home_pts, ball_pt, wr_pt, wr_label, wr_trail, cb_pt, cb_label, cb_trail, title]

def clear_dynamic():
    global arrow_artists, jersey_texts
    for a in arrow_artists:
        a.remove()
    arrow_artists = []
    for t in jersey_texts:
        t.remove()
    jersey_texts = []

def draw_team(frame_id, team_df, arrow_color, num_color):
    sub = team_df[team_df["frameId"] == frame_id]
    if sub.empty:
        return None, [], []
    pts = sub[["x","y"]].values
    arrows, texts = [], []
    for _, r in sub.iterrows():
        arr = ax.arrow(r["x"], r["y"], ARROW_SCALE*r["v_x"], ARROW_SCALE*r["v_y"],
                       head_width=0.8, head_length=0.8, length_includes_head=True,
                       fc=arrow_color, ec=arrow_color, alpha=0.9, lw=1.0, zorder=2)
        arrows.append(arr)
        num = "" if pd.isna(r.get("jerseyNumber")) else str(int(r["jerseyNumber"]))
        txt = ax.text(r["x"], r["y"], num, color=num_color, fontsize=9,
                      ha="center", va="center", zorder=5)
        texts.append(txt)
    return pts, texts, arrows

def _set_highlight(frame_id, by, val, pt_artist, label_artist, trail_x, trail_y, color="#000000"):
    if by not in track.columns or pd.isna(val):
        pt_artist.set_offsets(np.empty((0, 2))); label_artist.set_text("")
        return
    sub = track[(track["frameId"] == frame_id) & (track[by] == val)]
    if sub.empty:
        pt_artist.set_offsets(np.empty((0, 2))); label_artist.set_text("")
        return
    hx, hy = float(sub["x"].values[0]), float(sub["y"].values[0])
    pt_artist.set_offsets([[hx, hy]])
    # label
    label_txt = ""
    if "jerseyNumber" in sub.columns and pd.notna(sub["jerseyNumber"].values[0]):
        label_txt = str(int(sub["jerseyNumber"].values[0]))
    elif "displayName" in sub.columns and pd.notna(sub["displayName"].values[0]):
        label_txt = str(sub["displayName"].values[0])
    label_artist.set_position((hx, hy + 1.2))
    label_artist.set_text(label_txt)
    # trail
    trail_x.append(hx); trail_y.append(hy)
    return hx, hy  # optional return

def animate(frame_id):
    clear_dynamic()

    # Teams
    pts_away, txt_away, arr_away = draw_team(frame_id, away_df, away_1, away_2)
    if pts_away is not None:
        away_pts.set_offsets(pts_away); away_pts.set_edgecolors(away_2)

    pts_home, txt_home, arr_home = draw_team(frame_id, home_df, home_2, home_2)
    if pts_home is not None:
        home_pts.set_offsets(pts_home); home_pts.set_edgecolors(home_2)

    arrow_artists.extend(arr_away + arr_home)
    jersey_texts.extend(txt_away + txt_home)

    # Ball
    b = ball_df[ball_df["frameId"] == frame_id]
    if not b.empty:
        ball_pt.set_data(b["x"].values[0], b["y"].values[0])
    else:
        ball_pt.set_data([], [])

    # Highlights
    res_wr = _set_highlight(frame_id, HIGHLIGHT_WR_BY, HIGHLIGHT_WR_VAL, wr_pt, wr_label, wr_trail_x, wr_trail_y, color="#000000")
    wr_trail.set_data(wr_trail_x, wr_trail_y)

    res_cb = _set_highlight(frame_id, HIGHLIGHT_CB_BY, HIGHLIGHT_CB_VAL, cb_pt, cb_label, cb_trail_x, cb_trail_y, color="#cc0000")
    cb_trail.set_data(cb_trail_x, cb_trail_y)

    title.set_text(f"{title_txt}\nFrame: {frame_id}")
    return [away_pts, home_pts, ball_pt, wr_pt, wr_label, wr_trail, cb_pt, cb_label, cb_trail, title, *arrow_artists, *jersey_texts]

# Build animation
assert len(frame_ids) > 0, "frame_ids is empty"
anim = animation.FuncAnimation(
    fig, animate, init_func=init,
    frames=frame_ids, interval=100, blit=False, cache_frame_data=False
)

# Save or show
def _in_notebook():
    try:
        from IPython import get_ipython
        return "IPython" in globals() and get_ipython() is not None
    except Exception:
        return False

if _in_notebook():
    from IPython.display import HTML, display
    display(HTML(anim.to_jshtml()))
else:
    os.makedirs(os.path.dirname(OUT_GIF), exist_ok=True)
    writer = animation.PillowWriter(fps=10, metadata=dict(artist="you"))
    def _progress(i, n):
        if i == 0 or (i+1) == n or (i+1) % 10 == 0:
            print(f"Writing frame {i+1}/{n}...")
    anim.save(OUT_GIF, writer=writer, dpi=100, progress_callback=_progress)
    print(f"Saved {OUT_GIF} (from {os.path.basename(selected_week)})")
```
